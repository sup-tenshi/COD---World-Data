# Exploratory Data Analysis 

```python
df.isnull().sum() #showing the number of blank/null values in the data frame
```

```python
df1=df.drop('Code',axis='columns') # removing unwanted columns from the data frame
df1.isnull().sum()
```

```python
# Here we are converting the year to an object and Execution in to float for easy calculation
df2=df1
df2=df2.fillna(0)
df2.head(20)
```

```python
df2.isnull().sum()
```

```python
# Here we are converting the year to an object and Execution in to float for easy calculation
df2['Year'] = df['Year'].apply(str)
df2['Number of executions (Amnesty International)'] = pd.to_numeric(df2['Number of executions (Amnesty International)'],errors = 'coerce')
print(df2.info())
```

```python
num_col = df2._get_numeric_data().columns.tolist() #identifying the numberic columns 
print(num_col)
```

```python
# here we are trying to find ther totla 
df2['Total_Deaths_Per_Year'] = df2.sum(axis = 1)
#df2.loc['Total_Deaths']= df2.sum(numeric_only=True, axis=0)
df2 = df2.replace(np.nan, '', regex=True)
df2.head()
```
